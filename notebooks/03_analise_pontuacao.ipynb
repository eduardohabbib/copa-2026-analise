```python
# Importando as bibliotecas
import pandas as pd
import numpy as np

# Carregando os dados já limpos
df_ranking = pd.read_csv('../data/processed/ranking_limpo.csv')
df_resultados = pd.read_csv('../data/processed/resultados_limpos.csv')
df_copas = pd.read_csv('../data/processed/copas_limpas.csv')

print("✅ Dados limpos carregados com sucesso!")
print(f"\nRanking FIFA: {len(df_ranking)} seleções")
print(f"Resultados recentes: {len(df_resultados)} jogos")
print(f"Copas analisadas: {len(df_copas)} edições")
```

```python
# Calculando resultados para cada jogo
# Precisamos analisar cada seleção tanto como mandante quanto visitante

resultados_lista = []

for _, jogo in df_resultados.iterrows():
    home = jogo['home_team']
    away = jogo['away_team']
    gols_home = jogo['home_score']
    gols_away = jogo['away_score']

    if gols_home > gols_away:
        resultados_lista.append({'country_full': home, 'vitorias': 1, 'empates': 0, 'derrotas': 0})
        resultados_lista.append({'country_full': away, 'vitorias': 0, 'empates': 0, 'derrotas': 1})
    elif gols_home == gols_away:
        resultados_lista.append({'country_full': home, 'vitorias': 0, 'empates': 1, 'derrotas': 0})
        resultados_lista.append({'country_full': away, 'vitorias': 0, 'empates': 1, 'derrotas': 0})
    else:
        resultados_lista.append({'country_full': home, 'vitorias': 0, 'empates': 0, 'derrotas': 1})
        resultados_lista.append({'country_full': away, 'vitorias': 1, 'empates': 0, 'derrotas': 0})

# Convertendo para DataFrame e somando por seleção
df_aproveitamento = pd.DataFrame(resultados_lista)
df_aproveitamento = df_aproveitamento.groupby('country_full').sum().reset_index()

# Calculando aproveitamento em %
df_aproveitamento['jogos'] = df_aproveitamento['vitorias'] + df_aproveitamento['empates'] + df_aproveitamento['derrotas']
df_aproveitamento['pontos'] = df_aproveitamento['vitorias'] * 3 + df_aproveitamento['empates']
df_aproveitamento['aproveitamento'] = (df_aproveitamento['pontos'] / (df_aproveitamento['jogos'] * 3)) * 100

print("✅ Aproveitamento calculado!")
print(df_aproveitamento.sort_values('aproveitamento', ascending=False).head(10))
```

```python
# Filtrando apenas as 48 seleções classificadas
df_aproveitamento = df_aproveitamento[
    df_aproveitamento['country_full'].isin(selecoes_2026_corrigidas)
].reset_index(drop=True)

# Normalizando a nota de 0 a 100
aprv_max = df_aproveitamento['aproveitamento'].max()
df_aproveitamento['nota_aproveitamento'] = (df_aproveitamento['aproveitamento'] / aprv_max) * 100

print("✅ Critério 4 calculado!")
print(df_aproveitamento[['country_full', 'jogos', 'aproveitamento', 'nota_aproveitamento']]
      .sort_values('nota_aproveitamento', ascending=False).head(10))
```

```python
# Calculando gols marcados e sofridos por cada seleção
gols_lista = []

for _, jogo in df_resultados.iterrows():
    home = jogo['home_team']
    away = jogo['away_team']
    gols_home = jogo['home_score']
    gols_away = jogo['away_score']

    gols_lista.append({'country_full': home, 'gols_marcados': gols_home, 'gols_sofridos': gols_away})
    gols_lista.append({'country_full': away, 'gols_marcados': gols_away, 'gols_sofridos': gols_home})

# Convertendo para DataFrame e somando por seleção
df_gols = pd.DataFrame(gols_lista)
df_gols = df_gols.groupby('country_full').sum().reset_index()

# Calculando médias por jogo
df_gols['jogos'] = df_aproveitamento.set_index('country_full')['jogos']
df_gols = df_gols.merge(df_aproveitamento[['country_full', 'jogos']], on='country_full', how='inner')
df_gols['media_gols_marcados'] = df_gols['gols_marcados'] / df_gols['jogos']
df_gols['media_gols_sofridos'] = df_gols['gols_sofridos'] / df_gols['jogos']

# Filtrando só as 48 seleções
df_gols = df_gols[df_gols['country_full'].isin(selecoes_2026_corrigidas)].reset_index(drop=True)

print("✅ Gols calculados!")
print(df_gols[['country_full', 'media_gols_marcados', 'media_gols_sofridos']]
      .sort_values('media_gols_marcados', ascending=False).head(10))
```

```python
# Calculando gols marcados e sofridos por cada seleção
gols_lista = []

for _, jogo in df_resultados.iterrows():
    home = jogo['home_team']
    away = jogo['away_team']
    gols_home = jogo['home_score']
    gols_away = jogo['away_score']

    gols_lista.append({'country_full': home, 'gols_marcados': gols_home, 'gols_sofridos': gols_away})
    gols_lista.append({'country_full': away, 'gols_marcados': gols_away, 'gols_sofridos': gols_home})

# Convertendo para DataFrame e somando por seleção
df_gols = pd.DataFrame(gols_lista)
df_gols = df_gols.groupby('country_full').sum().reset_index()

# Cruzando com o número de jogos que já calculamos
df_gols = df_gols.merge(df_aproveitamento[['country_full', 'jogos']], on='country_full', how='inner')

# Calculando médias por jogo
df_gols['media_gols_marcados'] = df_gols['gols_marcados'] / df_gols['jogos']
df_gols['media_gols_sofridos'] = df_gols['gols_sofridos'] / df_gols['jogos']

print("✅ Gols calculados!")
print(df_gols[['country_full', 'media_gols_marcados', 'media_gols_sofridos']]
      .sort_values('media_gols_marcados', ascending=False).head(10))
```

```python
# Normalizando gols marcados (maior é melhor)
gols_marc_max = df_gols['media_gols_marcados'].max()
df_gols['nota_gols_marcados'] = (df_gols['media_gols_marcados'] / gols_marc_max) * 100

# Normalizando gols sofridos (menor é melhor — invertemos)
gols_sof_max = df_gols['media_gols_sofridos'].max()
gols_sof_min = df_gols['media_gols_sofridos'].min()
df_gols['nota_gols_sofridos'] = ((gols_sof_max - df_gols['media_gols_sofridos']) / (gols_sof_max - gols_sof_min)) * 100

print("✅ Critérios 5 e 6 calculados!")
print(df_gols[['country_full', 'nota_gols_marcados', 'nota_gols_sofridos']]
      .sort_values('nota_gols_marcados', ascending=False).head(10))
```

```python
# Juntando todos os critérios no DataFrame principal
df_final = df_copa.merge(df_aproveitamento[['country_full', 'nota_aproveitamento']], on='country_full', how='left')
df_final = df_final.merge(df_gols[['country_full', 'nota_gols_marcados', 'nota_gols_sofridos']], on='country_full', how='left')

# Preenchendo vazios com 0
df_final = df_final.fillna(0)

# Calculando pontuação final com os pesos definidos
df_final['pontuacao_final'] = (
    df_final['nota_ranking']          * 0.30 +
    df_final['nota_titulos']          * 0.10 +
    df_final['nota_campanha']         * 0.20 +
    df_final['nota_aproveitamento']   * 0.25 +
    df_final['nota_gols_marcados']    * 0.10 +
    df_final['nota_gols_sofridos']    * 0.05
)

# Ordenando pelo resultado final
df_final = df_final.sort_values('pontuacao_final', ascending=False).reset_index(drop=True)
df_final.index += 1

# Exibindo o ranking final
print("🏆 RANKING FINAL — FAVORITOS PARA A COPA 2026\n")
print(df_final[['country_full', 'pontuacao_final']].to_string())
```
