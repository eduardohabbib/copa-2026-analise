```python
# Importando as bibliotecas
import pandas as pd

# Carregando os arquivos
df_ranking = pd.read_csv('../data/raw/fifa_ranking-2024-06-20.csv')
df_resultados = pd.read_csv('../data/raw/results.csv')
df_copas = pd.read_csv('../data/raw/WorldCups.csv')
df_jogos = pd.read_csv('../data/raw/WorldCupMatches.csv')

print("✅ Arquivos carregados com sucesso!")
```

```python
# Verificando quais datas existem no ranking
print("Datas disponíveis no ranking FIFA:")
print(df_ranking['rank_date'].unique())
```

```python
# Filtrando apenas o ranking mais recente
df_ranking_limpo = df_ranking[df_ranking['rank_date'] == '2024-06-20']

# Removendo linhas sem rank
df_ranking_limpo = df_ranking_limpo.dropna(subset=['rank'])

# Resetando o índice
df_ranking_limpo = df_ranking_limpo.reset_index(drop=True)

# Verificando o resultado
print(f"Total de seleções no ranking: {len(df_ranking_limpo)}")
print(df_ranking_limpo.head(10))
```

```python
# Ordenando pelo rank e selecionando só as colunas que precisamos
df_ranking_limpo = df_ranking_limpo.sort_values('rank').reset_index(drop=True)

# Mantendo só as colunas úteis para o projeto
df_ranking_limpo = df_ranking_limpo[['rank', 'country_full', 'confederation', 'total_points']]

# Verificando o resultado
print(f"Total de seleções: {len(df_ranking_limpo)}")
print(df_ranking_limpo.head(10))
```

```python
# Convertendo a coluna de data para o formato de data do Python
df_resultados['date'] = pd.to_datetime(df_resultados['date'])

# Filtrando apenas jogos dos últimos 4 anos (2021 a 2024)
df_resultados_limpo = df_resultados[df_resultados['date'] >= '2021-01-01']

# Removendo jogos sem placar
df_resultados_limpo = df_resultados_limpo.dropna(subset=['home_score', 'away_score'])

# Resetando o índice
df_resultados_limpo = df_resultados_limpo.reset_index(drop=True)

# Verificando o resultado
print(f"Total de jogos recentes: {len(df_resultados_limpo)}")
print(df_resultados_limpo.head(5))
```

```python
# Filtrando apenas as últimas 3 Copas (2014, 2018, 2022)
df_copas_limpo = df_copas[df_copas['Year'] >= 2014]

# Selecionando apenas as colunas úteis
df_copas_limpo = df_copas_limpo[['Year', 'Country', 'Winner', 'Runners-Up', 'Third', 'Fourth']]

# Resetando o índice
df_copas_limpo = df_copas_limpo.reset_index(drop=True)

# Verificando o resultado
print(f"Total de Copas: {len(df_copas_limpo)}")
print(df_copas_limpo)
```

```python
# Adicionando manualmente as Copas que faltam no dataset
copas_faltando = pd.DataFrame({
    'Year': [2018, 2022],
    'Country': ['Russia', 'Qatar'],
    'Winner': ['France', 'Argentina'],
    'Runners-Up': ['Croatia', 'France'],
    'Third': ['Belgium', 'Croatia'],
    'Fourth': ['England', 'Morocco']
})

# Juntando com os dados que já temos
df_copas_limpo = pd.concat([df_copas_limpo, copas_faltando], ignore_index=True)

# Verificando o resultado
print(f"Total de Copas: {len(df_copas_limpo)}")
print(df_copas_limpo)
```

```python
# Salvando os dados limpos na pasta processed
df_ranking_limpo.to_csv('../data/processed/ranking_limpo.csv', index=False)
df_resultados_limpo.to_csv('../data/processed/resultados_limpos.csv', index=False)
df_copas_limpo.to_csv('../data/processed/copas_limpas.csv', index=False)

print("✅ Dados limpos salvos com sucesso!")
print("\nArquivos salvos em data/processed/:")
print("  - ranking_limpo.csv")
print("  - resultados_limpos.csv")
print("  - copas_limpas.csv")
```
