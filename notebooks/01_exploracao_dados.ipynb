```python
# Importando as bibliotecas
import pandas as pd

# Carregando os arquivos que vamos usar
df_ranking = pd.read_csv('../data/raw/fifa_ranking-2024-06-20.csv')
df_resultados = pd.read_csv('../data/raw/results.csv')
df_copas = pd.read_csv('../data/raw/WorldCups.csv')
df_jogos = pd.read_csv('../data/raw/WorldCupMatches.csv')

print("✅ Arquivos carregados com sucesso!")
```

```python
# Verificando o tamanho de cada tabela
print("📊 Ranking FIFA:")
print(f"   {df_ranking.shape[0]} linhas e {df_ranking.shape[1]} colunas")

print("\n📊 Resultados internacionais:")
print(f"   {df_resultados.shape[0]} linhas e {df_resultados.shape[1]} colunas")

print("\n📊 Copas do Mundo:")
print(f"   {df_copas.shape[0]} linhas e {df_copas.shape[1]} colunas")

print("\n📊 Jogos das Copas:")
print(f"   {df_jogos.shape[0]} linhas e {df_jogos.shape[1]} colunas")
```

```python
# Verificando as colunas de cada tabela
print("📊 Colunas do Ranking FIFA:")
print(df_ranking.columns.tolist())

print("\n📊 Colunas dos Resultados internacionais:")
print(df_resultados.columns.tolist())

print("\n📊 Colunas das Copas do Mundo:")
print(df_copas.columns.tolist())

print("\n📊 Colunas dos Jogos das Copas:")
print(df_jogos.columns.tolist())
```
